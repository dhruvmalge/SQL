```python
import tkinter as tk
import sqlite3
import mysql.connector as connection

mydb = connection.connect(
    host="localhost",
    username="root",
    password="Mysqlworkbench@8034",
    autocommit=True)

cursor = mydb.cursor()
cursor.execute("create database mytkinter")
cursor.execute("use mytkinter")
cursor.fetchall()



# Function to create a new database (if not exists)
def create_table():
    conn = connection.connect(
    host="localhost",
    username="root",
    password="Mysqlworkbench @ 8034",
    autocommit=True,
    database="mytkinter")
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS Records (
                    ID INTEGER PRIMARY KEY auto_increment,
                    Name varchar(255),
                    Age INT,
                    Email varchar(255))''')
    conn.commit()
    conn.close()

# Function to insert data into the database
def insert_data():
    name = name_entry.get()
    age = age_entry.get()
    email = email_entry.get()

    conn = connection.connect(
    host="localhost",
    username="root",
    password="Mysqlworkbench @ 8034",
    autocommit=True,
    database="mytkinter")
    cursor = conn.cursor()
    cursor.execute('''INSERT INTO Records (Name, Age, Email)
                      VALUES (%s, %s, %s)''', (name, age, email))
    conn.commit()
    conn.close()
    display_records()

# Function to update data in the database
def update_data():
    selected_id = id_entry.get()
    name = name_entry.get()
    age = age_entry.get()
    email = email_entry.get()

    conn = connection.connect(
    host="localhost",
    username="root",
    password="Mysqlworkbench @ 8034",
    autocommit=True,
    database="mytkinter")
    cursor = conn.cursor()
    cursor.execute('''UPDATE Records SET Name=%s, Age=%s, Email=%s WHERE ID=%s''',
                   (name, age, email, selected_id))
    conn.commit()
    conn.close()
    display_records()

# Function to delete data from the database
def delete_data():
    selected_id = id_entry.get()

    conn = connection.connect(
    host="localhost",
    username="root",
    password="Mysqlworkbench @ 8034",
    autocommit=True,
    database="mytkinter")
    cursor = conn.cursor()
    cursor.execute('''DELETE FROM Records WHERE ID=%s''', (selected_id,))
    conn.commit()
    conn.close()

# Function to display records in the GUI
def display_records():
    conn = connection.connect(
        host="localhost",
        username="root",
        password="Mysqlworkbench @ 8034",
        autocommit=True,
        database="mytkinter")
    cursor = conn.cursor()
    cursor.execute('''SELECT * FROM Records''')
    rows = cursor.fetchall()
    conn.close()

    # Clearing the display area
    display.delete(1.0, tk.END)

    # Displaying records in the text area
    for row in rows:
        display.insert(tk.END, f"ID: {row[0]}, Name: {row[1]}, Age: {row[2]}, Email: {row[3]}\n")

# Creating the GUI
root = tk.Tk()
root.title("Database Management")
root.geometry("500x500")

# Labels and entry fields for data input
tk.Label(root, text="ID:").grid(row=0, column=0)
id_entry = tk.Entry(root)
id_entry.grid(row=0, column=1)

tk.Label(root, text="Name:").grid(row=1, column=0)
name_entry = tk.Entry(root)
name_entry.grid(row=1, column=1)

tk.Label(root, text="Age:").grid(row=2, column=0)
age_entry = tk.Entry(root)
age_entry.grid(row=2, column=1)

tk.Label(root, text="Email:").grid(row=3, column=0)
email_entry = tk.Entry(root)
email_entry.grid(row=3, column=1)

# Buttons for CRUD operations
insert_button = tk.Button(root, text="Insert", command=insert_data)
insert_button.grid(row=4, column=0, pady=10)

update_button = tk.Button(root, text="Update", command=update_data)
update_button.grid(row=4, column=1, pady=10)

delete_button = tk.Button(root, text="Delete", command=delete_data)
delete_button.grid(row=4, column=2, pady=10)

# Text area to display records
display = tk.Text(root, height=10, width=40)
display.grid(row=5, columnspan=3)

# Initialize database and display existing records
create_table()
display_records()
delete_data()
            
root.mainloop()


cursor.execute("drop database mytkinter")
cursor.fetchall()

```
